```python
from neural_fly import utils
import matplotlib.pyplot as plt
import os
import numpy as np
from mpl_toolkits.mplot3d import Axes3D
```

```python
folder = '../data/training'
entries = os.listdir(folder)
print(entries)
filename = entries[1]

Data = utils.load_file(folder, filename)[0]
```

```python
t = Data['t']
p = Data['p']
p_d = Data['p_d']

x = p[:, 0]
y = p[:, 1]
z = p[:, 2]
xd = p_d[:, 0]
yd = p_d[:, 1]
zd = p_d[:, 2]
```

```python
# Create a figure with two subplots (one 3D and one 2D)
fig, axes = plt.subplots(1, 2, figsize=(12, 5), width_ratios=(1,2))

# First subplot (3D plot)
ax = axes[0]
ax.remove()
ax = fig.add_subplot(121, projection='3d')  # Convert to 3D manually
ax.plot(x, y, z, label='Actual', color='r', linewidth=1)
ax.plot(xd, yd, zd, label='Desired', color='r', linewidth=1, alpha=0.3)
ax.set_title('hover pid with fa=2')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
ax.set_xlim([-2,2])
ax.set_zlim([-2,2])
ax.set_ylim([-2,2])
ax.legend()

# Second subplot (2D plot)
ax = axes[1]
ax.plot(t, x, label='x', color='gray')
ax.plot(t, xd, label='xd', color='gray', alpha=0.3)
ax.plot(t, y, label='y', color='orange')
ax.plot(t, yd, label='yd', color='orange', alpha=0.3)
ax.plot(t, z, label='z', color='teal')
ax.plot(t, zd, label='zd', color='teal', alpha=0.3)
ax.set_xlabel('t')
ax.set_ylabel('x, y, z')
ax.legend()
ax.grid()

# Adjust layout
plt.tight_layout()
plt.subplots_adjust(wspace=0.3)
plt.show()
```
